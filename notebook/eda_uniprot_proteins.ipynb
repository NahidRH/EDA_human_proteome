```python
project_path = "/content/drive/MyDrive/GitHub/eda-human-proteome"
data_path = project_path + "/data/sample_uniprot_100.tsv"
```

```python
# Display basic information

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Set default plot style
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Define path to the UniProt TSV file
data_path = data_path

# Load the dataset
df = pd.read_csv(data_path, sep='\t')

# Display basic information
print("Dataset shape:", df.shape)
print("Available columns:", df.columns.tolist())

# Show the first few rows
df.head()
```

```python
# Plot the distribution of protein sequence lengths
sns.histplot(df['Length'], bins=50, kde=True)
plt.title("Distribution of Protein Sequence Lengths")
plt.xlabel("Length (Amino Acids)")
plt.ylabel("Number of Proteins")
plt.show()
```

```python
# Plot the distribution of protein molecular masses
sns.histplot(df['Mass'], bins=50, kde=True, color='orange')
plt.title("Distribution of Molecular Weights")
plt.xlabel("Mass (Daltons)")
plt.ylabel("Number of Proteins")
plt.show()
```

```python
#Top 10 Subcellular Locations

# Handle missing values
df['Subcellular location [CC]'] = df['Subcellular location [CC]'].fillna('')

# Split and explode subcellular locations
locations = df['Subcellular location [CC]'].fillna('').str.split(';').explode().str.strip()
top_locations = locations.value_counts().head(10)

# Optional: truncate long labels
top_locations.index = [loc[:40] + '...' if len(loc) > 40 else loc for loc in top_locations.index]

fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(y=top_locations.index, x=top_locations.values, ax=ax, color='skyblue')
ax.set_title("Top 10 Subcellular Locations")
ax.set_xlabel("Number of Proteins")
ax.set_ylabel("Location")
plt.tight_layout()
plt.show()
```

```python
# Top 20 Functional Keywords in Human Proteins

# Handle missing values
df['Keywords'] = df['Keywords'].fillna('')

# Split and explode keywords (each row may have multiple keywords separated by ';')
keywords = df['Keywords'].str.split(';').explode().str.strip()

# Count most frequent keywords
top_keywords = keywords.value_counts().head(20)

# Create a DataFrame for plotting with hue
df_keywords = pd.DataFrame({
    "Keyword": top_keywords.index,
    "Count": top_keywords.values
})

# Plot with hue to enable palette safely
fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(data=df_keywords, y="Keyword", x="Count", hue="Keyword", dodge=False, palette="viridis", ax=ax)

ax.set_title("Top 20 Functional Keywords in Human Proteins")
ax.set_xlabel("Number of Proteins")
ax.set_ylabel("Functional Keyword")
ax.legend_.remove()  # remove the color legend (not needed)
plt.tight_layout()
plt.show()
```
